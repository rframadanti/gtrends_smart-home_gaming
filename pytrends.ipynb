```python
import pandas as pd
import numpy as np
```

```python
from pytrends.request import TrendReq

pytrends = TrendReq(hl='en-US', tz=360)
```

Build Payload & keywords list

```python
def gtrends(keyword_filedirectory, column_name, keyword_relative, naming_optional):
    category = pd.read_csv(keyword_filedirectory)
    full_list = category[column_name].values.tolist()
    #iterate every 4 item in list plus a keyword as the relative comparison
    i = 0
    while i < len(full_list):
        l = full_list[i:(i+4)]
        l.append(keyword_relative)
        pytrends.build_payload(l, cat=0, timeframe='all', geo='', gprop='')
        df_time = pytrends.interest_over_time()
        df_time.reset_index(inplace=True)
        df_time_name = "gtrends_overtime"+str(naming_optional)+str((i+4)//4)+".csv"
        df_time.to_csv(df_time_name, index = False)
        df_region = pytrends.interest_by_region(resolution='COUNTRY', inc_low_vol=True, inc_geo_code=False)
        df_region.reset_index(inplace=True)
        df_region_name = "gtrends_region"+str(naming_optional)+str((i+4)//4)+".csv"
        df_region.to_csv(df_region_name, index = False)
        i += 4
```

```python
def combine_wbase(base_name, n_file, baseline, filename):
    df1 = pd.read_csv(base_name+str(1)+".csv")
    for i in range(n_file-1):
        df2 = pd.read_csv(base_name+str(i+2)+".csv")
        df1 = pd.concat([df1, df2], axis=1, sort=False)
    base = pd.read_csv(baseline)
    df1 = pd.concat([df1, base], axis=1, sort=False)
    df_name = filename
    df1.to_csv(df_name, index = False)
```

```python
def combine(base_name, n_file, filename):
    df1 = pd.read_csv(base_name+str(1)+".csv")
    for i in range(n_file-1):
        df2 = pd.read_csv(base_name+str(i+2)+".csv")
        df1 = pd.concat([df1, df2], axis=1, sort=False)
    df_name = filename
    df1.to_csv(df_name, index = False)
```

```python
combine("gtrends_region-v0_", 18, "gtrends_region_combine.csv")
df1 = pd.read_csv("gtrends_region_combine.csv")
df1
```

```python
def tidydata(df, nf, col1, base_kw, filename, col2, pairs=4):
    df1 = pd.read_csv(df)
    n_file = nf
    for i in range(n_file-1):    
        df1 = df1.drop(columns=""+str(col1)+str(i+1)+"")
        
    row = (len(df1.index)-1)
    l_check_more = []
    l_check_less = []
    for i in range(n_file-1):
        if df1[""+str(base_kw)+"."+str(i+1)+""][row] == df1[base_kw][row]:
            pass
        elif df1[""+str(base_kw)+"."+str(i+1)+""][row] < df1[base_kw][row]:
            l_check_less.append(i+1)
        elif df1[""+str(base_kw)+"."+str(i+1)+""][row] > df1[base_kw][row]:
            l_check_more.append(i+1)

    l = l_check_more + l_check_less

    for i in l:
        k = df1.columns.get_loc(""+str(base_kw)+"."+str(i)+"")
        for n in range(len(df1.index)):
            for m in range(pairs):
                df1.iloc[n,(k-m-1)] = (df1.iloc[n,(k-m-1)] * (df1[base_kw][n]/df1[""+str(base_kw)+"."+str(i)+""][n]))
    
    df1.fillna(0, inplace=True)
    for i in range(n_file-1):    
        df1 = df1.drop(columns=""+str(base_kw)+"."+str(i+1)+"")
    
    df2=pd.melt(df1,id_vars=[col2],var_name='keywords', value_name='relative_popularity')
    df2.to_csv(filename, index = False)

tidydata("gtrends_region_combine.csv", 18, "geoName.", "Chromecast", "gtrends_region_combine-v1.csv", 'geoName')
df3 = pd.read_csv("gtrends_region_combine-v1.csv")
df3
```
